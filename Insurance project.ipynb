```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix,mean_squared_error,r2_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('Untitled Folder/insurance.csv')
```

```python
plt.figure(figsize=(12,5))

sns.boxplot(x = df['bmi'])
plt.title("Box plot of BMI")
```

```python
q1 = df['bmi'].quantile(0.25)
q3 = df['bmi'].quantile(0.75)

IQR = q3 - q1

L_bound = q1-1.5*IQR
U_bound = q3+1.5*IQR

print(f"The lower bound is {L_bound} and the upper bound is {U_bound}")
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df_cleaned = df[(df['bmi']>=L_bound) & (df['bmi']<= U_bound)]
```

```python
plt.figure(figsize=(7,3.5))

sns.histplot( df_cleaned['bmi'] , kde = True)
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(7,3.5))
sns.histplot( df['bmi'] , kde = True)
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.show()
```

```python
df_cleaned
```

```python
# One Hot encoding
df_cleaned = pd.get_dummies(df_cleaned,drop_first = True)
df_cleaned
```

```python
df_cleaned.corr()
```

```python
plt.figure(figsize = (12,8))
sns.heatmap(df_cleaned.corr(),annot=True)
```

```python
x = df_cleaned[['age', 'bmi','smoker_yes']]
y = df_cleaned['charges']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=23)
lr_model = LinearRegression()
lr_model.fit(x_train,y_train)
```

```python
lr_pred = lr_model.predict(x_test)
```

```python
y_train_pred = lr_model.predict(x_train)
```

```python
print(f' Testing results : {r2_score(y_test,lr_pred)}')
print(f' Training results : {r2_score(y_train,y_train_pred)}')
```
